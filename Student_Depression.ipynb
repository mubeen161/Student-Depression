```python
X=df.drop('Depression',axis=1)
Y=df['Depression']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
```

```python
ypred=rf.predict(X_test)
```

```python

# Function to predict depression based on user input
def predict_depression():
    # Create a dictionary to store user input
    user_input = {}

    # Get user input for each feature
    features = ['Gender', 'Age', 'Academic Pressure', 'Study Satisfaction', 'Sleep Duration',
                'Dietary Habits', 'Have you ever had suicidal thoughts ?', 'Study Hours',
                'Financial Stress', 'Family History of Mental Illness']

    for feature in features:
        if df[feature].dtype == 'object':  # Categorical features
          unique_values = df[feature].unique()
          print(f"Enter {feature} (Options: {unique_values}):")
          while True:
              user_value = input()
              if user_value in [str(x) for x in unique_values]:
                user_input[feature] = user_value
                break
              else:
                print(f"Invalid input. Please choose from: {unique_values}")
        else:
          while True:
              try:
                  user_value = int(input(f"Enter {feature}: "))
                  user_input[feature] = user_value
                  break
              except ValueError:
                  print("Invalid input. Please enter an integer.")

    # Create a DataFrame from user input
    input_df = pd.DataFrame([user_input])

    # Apply Label Encoding to user input
    lb = LabelEncoder()
    for col in input_df.columns:
        if input_df[col].dtype == 'object':
            input_df[col] = lb.fit_transform(input_df[col])

        #Handle potential type mismatch for numerical columns in user input vs. trained data
        elif input_df[col].dtype != df[col].dtype:
          input_df[col] = input_df[col].astype(df[col].dtype)

    # Make the prediction
    prediction = lr.predict(input_df)

    # Print the prediction
    if prediction[0] == 0:
      print("Prediction: No Depression")
    else:
      print("Prediction: Depression")
# Call the function to start the prediction process
predict_depression()
```
